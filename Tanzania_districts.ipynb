## Imports Required Libraries

```python
%matplotlib inline
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
```

## Loading Datasets

```python
Tanzania_districts_df = gpd.read_file('../Datasets/Tanzania Districts Administrative Boundaries shp/Districts and TC as 2020.shp')
Tanzania_districts_df.head(10)
```

## List of all Tanzania Districts

```python
Tanzania_districts_df['Districts'] = Tanzania_districts_df['NewDist20']
Tanzania_districts_df['Districts'].unique()
```

## Tanzania Districts 

```python
Tanzania_Districts = Tanzania_districts_df['Districts'].unique()
Tanzania_districts_df = Tanzania_districts_df[Tanzania_districts_df.Districts.isin(Tanzania_Districts)]
fig, ax = plt.subplots(figsize=(10,10))
ax.axis('off')
for idx, row in Tanzania_districts_df.iterrows():
    centroid = row['geometry'].centroid
    ax.annotate(row['Districts'], (centroid.x, centroid.y), textcoords='offset points', xytext=(0,10), ha='center', fontsize=6, color='black')
Tanzania_districts_df.plot(ax=ax, column='Districts')
#plt.savefig('../Outputs/Tanzania Zones/Tanzania Districts.png', dpi=300, bbox_inches='tight')
```

## Tanzania Individual Districts Map

```python
Tanzania_Districts = Tanzania_districts_df['Districts'].unique()
for Districts in Tanzania_Districts:
    Districts_df = Tanzania_districts_df[Tanzania_districts_df.Districts.eq(Districts)]
    fig, ax = plt.subplots(figsize=(10,10))
    ax.axis('off')
    for idx, row in Districts_df.iterrows():
        centroid = row['geometry'].centroid
        ax.annotate(row['Districts'], (centroid.x, centroid.y), textcoords='offset points', 
                    xytext=(0,10), ha='center', fontweight='bold', fontsize=10, color='white')
    Districts_df.plot(ax=ax, color='#263A47', edgecolor='black')
    plt.savefig(f'../Outputs/Tanzania Districts Administrative Boundaries/{Districts} Map.png', dpi=300, bbox_inches='tight')
    plt.close()
print('Plots Saved Succesfully')
```
